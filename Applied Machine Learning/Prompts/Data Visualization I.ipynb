### Import libraries

```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

Run the line below to see plots in Jupyter notebook

```python
%matplotlib inline
```

### Import data sales.csv

```python
df = pd.read_csv('sales.csv')
df['date'] = df.date.astype('datetime64[ns]')
```

```python
df.shape
```

```python
df.head()
```

### Matplotlib Commands

```python
plt.figure(figsize=(15,6))
plt.plot(df.week, df.price)
plt.xlabel('Week') 
plt.ylabel('Price ($)') 
plt.title('Weekly Prices')
plt.show() 
```

```python
plt.figure(figsize=(15,6)) 
plt.plot(df.date, df.quantity) 
plt.xlabel('Date')
plt.ylabel('Quantity') 
plt.title('Weekly Quantities') 
plt.show() 
```

### Creating Multiplots on Same Canvas

```python
plt.figure(figsize=(15,6)) 
plt.subplot(2,1,1)
plt.plot(df.date, df.price) 
plt.ylabel('Price ($)') 
plt.xticks(rotation=45) 

plt.subplot(2,1,2)
plt.plot(df.date, df.quantity) 
plt.xlabel('Date') 
plt.ylabel('Quantity')
plt.xticks(rotation=45) 

plt.tight_layout() 
```

The above are methods (functions) for 'pyplot'. See here for other functions you can use: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.html

# Matplotlib Object Oriented Method


Object Oriented method is the formal and convenient way to create figure objects and call methods or attributes of that object. 

First, we create a figure instance, and add axes to that figure.

```python
# Create figure with 1 subplot (name the figure 'fig', and its axes 'ax')
fig, ax = plt.subplots(figsize=(15,6))

ax.plot(df.date, df.price, label='price label') 
ax.plot(df.date, df.cost, label='cost') 
ax.set_xlabel('Date') 
ax.set_ylabel('Dollar ($)')
ax.set_title('Weekly Prices and Costs')

# Add a legend in a location 
ax.legend(loc=3) 
```

The above are methods (functions) for 'axes'. See here for other methods you can use: https://matplotlib.org/3.1.1/api/axes_api.html#matplotlib.axes.Axes

Similarly, see here for functions you can use for 'figure': 
https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.figure.Figure.html

### Saving figures
Matplotlib can save figures in formats such as PNG, JPG, EPS, SVG, PGF and PDF. 

```python
fig.savefig("fig1.png")
fig.savefig("fig1.jpeg")
```

Here we can also set the DPI (dots per inch):

```python
fig.savefig("fig1_dpi_200.png", dpi=200)
fig.savefig("fig1_dpi_500.jpeg", dpi=500)
```

## Setting colors, linewidths, linetypes

```python
fig, ax = plt.subplots(figsize=(15,6))

ax.plot(df.date, df.price, 'b.--')
ax.plot(df.date, df.cost, 'g--') 
ax.set_xlabel('Date') 
ax.set_ylabel('Dollar ($)')
ax.set_title('Weekly Prices and Costs')
```

Some other examples for line styles: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D


### Line and marker styles

To change the line width, we can use the `linewidth` or `lw` keyword argument. The line style can be selected using the `linestyle` or `ls` keyword arguments:

```python
fig, ax = plt.subplots(figsize=(15,6))

ax.plot(df.date, df.price, linewidth=0.5)
ax.plot(df.date, df.cost, linewidth=4) 
ax.set_xlabel('Date') 
ax.set_ylabel('Dollar ($)')
ax.set_title('Weekly Prices and Costs')
```

```python
fig, ax = plt.subplots(figsize=(15,6))

ax.plot(df.date, df.price, marker='+', markersize=15) 
ax.plot(df.date, df.cost, marker='o', markersize=4) 
ax.set_xlabel('Date') 
ax.set_ylabel('Dollar ($)')
ax.set_title('Weekly Prices and Costs')
```
